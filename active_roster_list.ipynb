```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_current_NBA_team_rosters"
```

```python
team_list = []

tables = pd.read_html(url)

for team in tables:
    try:
        roster = team.iloc[2:-1, 2]
        roster_list = roster.to_list()
        team_list.append(roster_list)
    except:
        pass


team_list
# first = tables[0]
# test = first.iloc[2:-1, 2]
# print(test.to_list())
```

```python
player_list = []
for team in team_list:
    for player in team:
        try:
            player_list.append(player.split("\xa0")[0])
        except:
            player_list.append(player)
                
```

```python
players = []
for player in player_list:
    try:
        name = player.split(", ")
        first_name = name[1]
        last_name = name[0]
        players.append(first_name + " " + last_name)
    except:
        players.append(player)
```
