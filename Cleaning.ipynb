```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect

import pandas as pd
import numpy as np
```

```python
file_path = "missing.csv"
mis_df = pd.read_csv(file_path)
mis_df = mis_df.copy()
mis_df.head()
```

```python
to_remove = [c for c in mis_df.columns if "Unnamed" in c]
# to_remove

mis_df.drop(to_remove, axis=1, inplace= True)

to_remove2 = [d for d in mis_df.columns if "_x" in d]

mis_df.drop(to_remove2, axis=1, inplace= True)

mis_df.head()
```

```python
file_path = "HOF_stats.csv"
hof_df = pd.read_csv(file_path)
hof_df = hof_df.copy()
hof_df.head()

# if the season doesn't contain a -, delete the row
```

```python
# for col in hof_df.columns:
#     print(col)
to_remove = [c for c in hof_df.columns if "Unnamed" in c]
# to_remove

hof_df.drop(to_remove, axis=1, inplace= True)

to_remove2 = [d for d in hof_df.columns if "_x" in d]

hof_df.drop(to_remove2, axis=1, inplace= True)

hof_df.head()

# for col in df.columns:
#     print(col)
```

```python
file_path = "Active_roster_stats.csv"
ars_df = pd.read_csv(file_path)
ars_df = ars_df.copy()
ars_df.head()
```

```python
# for col in ars_df.columns:
#     print(col)
to_remove = [c for c in ars_df.columns if "Unnamed" in c]
# to_remove

ars_df.drop(to_remove, axis=1, inplace= True)

to_remove2 = [d for d in ars_df.columns if "_x" in d]

ars_df.drop(to_remove2, axis=1, inplace= True)

# players = ars_df["Player"]
ars_df.head()
# for col in df.columns:
#     print(col)
```

```python
file_path = "MVP_stats.csv"
mvp_df = pd.read_csv(file_path)
mvp_df = mvp_df.copy()
mvp_df.head()
```

```python
to_remove = [c for c in mvp_df.columns if "Unnamed" in c]
# to_remove

mvp_df.drop(to_remove, axis=1, inplace= True)

to_remove2 = [d for d in mvp_df.columns if "_x" in d]

mvp_df.drop(to_remove2, axis=1, inplace= True)

# players = mvp_df["Player"]
mvp_df.head()
```

```python
master_df = hof_df.append(ars_df).append(mvp_df).append(mis_df)

master_df.head()
```

```python
master_df = master_df.drop_duplicates(subset=['Player', 'Season'], keep='first')
master_df = master_df.dropna(subset = ["Age_y"])
master_df.info()
```

```python
to_rename = {"Age_y":"Age", "G_y":"G",
             "Lg_y":"Lg", "MP_y": "MP",
             "Pos_y": "Pos", "Tm_y":"Tm"}
master_df = master_df.rename(columns=to_rename)
master_df.head()
```

```python
master_df["GT"] = master_df.sum(numeric_only=True, axis=1)

career_best_df = master_df.sort_values('GT').drop_duplicates(['Player'], keep='last')


```
